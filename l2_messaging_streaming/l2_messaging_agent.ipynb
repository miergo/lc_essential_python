```python
from dotenv import load_dotenv
from langchain.agents import create_agent
from langchain_core.messages import HumanMessage
from langchain.tools import tool
from langchain_ollama import ChatOllama


load_dotenv()
```

```python
SYSTEM_PROMPT=system_prompt="""You are a polish comedian, 

- Think step-by-step for the joke to make sense.
- reply in polish only but add some english inbetween. 
- Feel free to use polish curse words to make the jokes funnier.
- Use Polish banter for better delivery

"""
```

```python
agent1 = create_agent(
    model = ChatOllama(
        model="qwen3.5:9b",
        reasoning=True,
    ),
    system_prompt=SYSTEM_PROMPT
)


```

```python
human_msg = "How is the weather in Opole?"


for token, metadata in agent1.stream(
    {"messages": [{"role": "user", "content": human_msg}]},
    #stream_mode="values",
    stream_mode="messages", #prints each token after another
):
    print(f"{token.content}", end="")
```

```python
for step in agent1.stream(
    {"messages": [{"role": "user", "content": human_msg}]},
    stream_mode="values",
    #stream_mode="messages", #prints each token after another
):
    step['messages'][-1].pretty_print()
```
